#### Converting PIL image to RGB, Graysclae or black n white

Normaly PIL loads an image in RGB format (RED, GREEN, BLUE) format. it will have 3 channels. But a PNG may have an additional 4th channel too.


To load image in grayscale using PIL

```python
from PIL import Image
img = Image.load(im_path)
img.convert("RGB") # to RGB
img.convert("L") # to grayscale
img.convert("1") # to black & white
```

````python
from PIL import Image

im_path = "shrek.jpg"
img = Image.open(im_path)

rgb_img = img.convert("RGB") # if it is in RGBA, then it will remove
                             # its 4th channel and resulting will be in RGB
                             # if it is in grayscale, it will convert grascale value
                             # to equivalent RGB, which will result same grayscale color
display(rgb_img)

grayscale_img = img.convert("L")
display(grayscale_img)

bw_img = img.convert("1")
display(bw_img)
````

#### CV2 image channels and conversion

Normaly cv2 loads an image into BGR format, while all other libraries loads image into RGB. In cv2 image Red is replaced with Blue.

```python
import cv2
from PIL import Image

cv2_img = cv2.imread(im_path)
display(Image.fromarray(cv2_img))
```

here channels are switched. We can convert into RGB using following code

```python
converted_img = cv2.cvtColor(cv2_img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(converted_img))
```

into GRAY SCALE

```python
converted_img = cv2.cvtColor(cv2_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(converted_img))
```

````python
import cv2
from PIL import Image

im_path = "shrek.jpg"
cv2_img = cv2.imread(im_path)
display(Image.fromarray(cv2_img))

converted_img = cv2.cvtColor(cv2_img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(converted_img))

# converting it into grayscale
converted_img = cv2.cvtColor(cv2_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(converted_img))

# loading an image into grayscale mode
cv2_img = cv2.imread(im_path, 0)
display(Image.fromarray(cv2_img))
````

#### Resize CV2 Image

Write code to resize CV2 image

````python
import cv2
import matplotlib.pyplot as plt

im_path = "shrek.jpg"
cv2_img = cv2.imread(im_path)

resized_img = cv2.resize(cv2_img, (300,300))

plt.imshow(resized_img)
plt.show()
````

#### Rotate CV2 Image

Write code to crop CV2 image


````python
import cv2
import matplotlib.pyplot as plt

im_path = "shrek.jpg"
cv2_img = cv2.imread(im_path)

rotated_img = cv2.rotate(cv2_img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated_img)
plt.show()
````

#### Crop CV2 Image

Write code to crop CV2 image

````python
import cv2
import matplotlib.pyplot as plt

im_path = "shrek.jpg"
cv2_img = cv2.imread(im_path)

croped_img = cv2_img[0:100, 0:100]   # there is no function cv2.crop in cv2 so we have to use [start_y:end_y, start_x:end_x]

plt.imshow(croped_img)
plt.show()
````
